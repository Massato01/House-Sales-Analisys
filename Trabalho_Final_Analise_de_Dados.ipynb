```
# Criar um DataFrame (df_faltantes) contendo a quantidade total de elementos NaN em cada coluna em uma lista ordenada do maior para o menor
df_faltantes = df.isna().sum().sort_values(ascending = False)
df_faltantes.head(22)
```

```
# Exibir um gráfico de barras das quantidades de itens NaN das colunas com valor diferente de 0
# Os dados devem estar do maior para o menor

# SEABORN
# sns.barplot(x = df_faltantes.iloc[0:19], y = df_faltantes.index[0:19]);

# PLOTLY
df_faltantes.iloc[0:19].plot(kind = 'bar',
                             title = 'Quantidade de itens NaN das colunas com valor diferente de 0')
```

```
# Calcular a porcetagem de itens faltantes com relação a quantidade de entradas
df_original = df.count().sort_values(ascending = True)

# Temos 1460 entradas para 80 colunas
print(f'FILLED VALUES:\n{df_original.head(19)}')
print(f'\nMISSING VALUES:\n{df_faltantes.head(19)}')

# Exibir as porcentagens em ordem decrescente
dif = df_faltantes / 1460
percent = dif * 100
print(f'\nMISSING VALUES PERCENTAGE:\n{(percent).head(19)}')

# Exibir o resultado em um gráfico de barras em ordem decrescente
percent[0:19].plot(kind = 'bar',
                   title = 'Porcentagem de itens faltantes')
```

```
# Criar um novo DataFrame (df_limpo) sem as colunas com mais de 90% de itens faltantes
df_limpo = df.copy()
for columns in df.iloc[0:19]:
  if percent[columns] > 90:
    df_limpo = df_limpo.drop([columns], axis = 1)
```

```
# Determine o preço médio de venda das casas
print(f'AVERAGE PRICE FOR HOUSES ON SALE: US${df_limpo.SalePrice.mean():.2f}')
```

```
# Exibir o gráfico de linha do preço de venda pelo ano de construção com uma linha indicando o valor médio de venda
# SalePrice ~ YearBuilt
'''SALE PRICE FOR BUILT YEAR'''

# INDEPENDENT VARIABLE
X = df_limpo.iloc[:, 76].values

# DEPENDENT VARIABLE
y = df_limpo.iloc[:, 17].values

correlacao = np.corrcoef(X, y)

print(f'SalePrice: {X}\n')
print(f'YearBuilt: {y}\n')
print(f'Correlacao: {correlacao}')
```

```
# -> Formatando X
X = X.reshape(-1, 1)

# Criando o modelo de Regressao Linear
from sklearn.linear_model import LinearRegression

# Treinando o modelo
modelo = LinearRegression()
modelo.fit(X, y)
```

```
# Plotando o grafico
plt.figure(figsize = (15, 10))
plt.scatter(X, y, color = 'royalblue', alpha = 0.65, s = 70, facecolor = 'white')
plt.plot(X, modelo.predict(X), color = 'darkblue')
plt.xlabel('SalePrice')
plt.ylabel('YearBuilt')
plt.show()

#PLOTANDO COM PLOTLY
import plotly.express as px
df = px.data.tips()
fig = px.scatter(df_limpo, x='SalePrice', y='YearBuilt', opacity=0.65,
                 trendline = 'ols', trendline_color_override='darkblue')

fig.show()
```

```
# Filtrando apenas as casas antigas ( < 1880)
before_80 = []
for year in df_limpo['YearBuilt']:
  if year < 1880:
    before_80.append('Before80')
  else:
    before_80.append('After80')

df_limpo['CasasAntigas'] = before_80
```

```
# Média de preço de casas antigas
df_limpo.loc[df_limpo['CasasAntigas'] == 'Before80'].mean()
```

```
# Na coluna "LotFrontage" substituir os valores faltantes pela média da coluna
df_limpo = df_limpo.fillna(round(df_limpo.LotFrontage.mean()))
```

```
# Os valores nas colunas "LotFrontage" e "LotArea" estão em ft e ft^2 converter para m e m^2
# 1 ft = 0.3048 m
# 1 ft^2 = 0.092903 m^2

for line in df_limpo['LotFrontage']: # O primeiro valor esta sendo convertido 2 vezes!!
  conv = line * 0.3048
  df_limpo.replace(line, conv, inplace = True)


for line in df_limpo['LotArea']:
  conv = line * 0.3048
  df_limpo.replace(line, conv, inplace = True)

print(df_limpo.LotFrontage.head()), print()
print(df_limpo.LotArea.head())
```

```
# Quais as quantidades únicas de cômodos (TotRmsAbvGrd) existem nos imóveis disponíveis
df_limpo.TotRmsAbvGrd.value_counts()
```

```
# Quantos tipos de garagem existem? E quantos imóveis de cada tipo existem?
# GarageType and SaleType
print(f'GARAGE TYPE:\n{df_limpo.GarageType.value_counts()}')
print(f'\nSALE TYPE:\n{df_limpo.SaleType.value_counts()}')
```

```
# Quantos imóveis foram construídos em cada ano?
# SaleType ~ YearBuilt

for value in df_limpo['YearBuilt']:
  if value < 1000:
    df_limpo.replace(value, df_limpo.YearBuilt.mean(), inplace = True)

plt.figure(figsize = (15, 10))
plt.bar(df_limpo.YearBuilt, df_limpo.SalePrice, color = 'royalblue')
plt.xlabel('YearBuilt')
plt.ylabel('SalePrice')
plt.xticks()

plt.tight_layout()
plt.show()
```

```
# Agrupar os imóveis pelo ano que foram construídos, reformados e vendidos
# YearBuilt, YearRemodAdd, YrSold

df_limpo.YearBuilt = df_limpo.YearBuilt.astype(int)
df_limpo.YearRemodAdd = df_limpo.YearRemodAdd.astype(int)
df_limpo.YrSold = df_limpo.YrSold.astype(int)

immobile = df_limpo.groupby(['YearBuilt', 'YearRemodAdd', 'YrSold']).sum()
immobile.sort_values(by = 'YrSold').head(10)
```
